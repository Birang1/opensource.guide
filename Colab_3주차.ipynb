```python
# 위에서 한 내용을 코드에 따라 자동화 시키기

# 이 내용은 필요 없음. corp_code = df_listed[df_listed['corp_name'] == '카카오'].iloc[0,0]

def get_shareholders(corp_code):

  data = dart_fss.api.info.hyslr_sttus(corp_code, '2021', '11011') # https://dart-fss.readthedocs.io/en/latest/dart_api.html#module-dart_fss.api.finance 최대주주현황 코드 있음

  df= pd.DataFrame(data['list']) # DataFrame안에는 () 소괄호로 코딩 진행

  # dataframe안에 있는 column들을 모아서 dataframe으로 만든다!
  df = df[['corp_name','nm','relate','bsis_posesn_stock_qota_rt','trmend_posesn_stock_qota_rt','rm']]

  df.columns = ['기업명','이름','관계','기초지분율','기말지분율','비고']

  df = df[df['관계'].notnull()] # 관계 : NaN인 사람들을 지운것을 dataframe 으로 한다.

  # 기초지분율과 기말지분율의 값들을 모두 숫자로 바꾼다.

  df['기초지분율'] = pd.to_numeric(df['기초지분율'])
  df['기말지분율'] = pd.to_numeric(df['기말지분율'])

  df

  # 기초지분율의 높은 3사람을 본다.
  return df.sort_values(by = '기초지분율', ascending = False).head(3)

```

```python
# 위의 내용을 실행 하고 난뒤 진행해야 뒤의 for 구문안의 get_shareholders 함수가 돌아갈것임.
# df_listed 데이터프레임에서 10개의 내용을 랜덤으로 샘플링 하는것임.
# corp_code라는 부분을 랜덤으로 10개를 샘플링 해라.
df_listed.sample(10)['corp_code']

# 이를 list로 받고 corp_codes라고 변수로 두자
corp_codes=list(df_listed.sample(10)['corp_code'])

corp_codes

# 데이터 프레임을 이어붙힐 list를 dfs라 두고
dfs = []
for corp_code in corp_codes:
  try:
    df = get_shareholders(corp_code) # 최대 주주 지분율에 대한 함수를 실행 시키는데,
    dfs.append(df) # 각 코드마다 나오는 dataframe을 이어 붙힌다.
    
  except:
    print(f'error-{corp_code}') 

df_result = pd.concat(dfs) # 데이터 프레임을 이어 붙힌 data frame 결과를 df_result라고 하자
df_result['증감'] = df_result['기말지분율'] - df_result['기초지분율'] # 기말지분율에서 기초지분율의 각 값을 뺀 상태를 '증감'이라고 하자.

df_result.sort_values(by = '증감', ascending=False)
```

```python
# 위의 것을 함수로 만들기

def get_profit(name):
  corp_code = df_listed[df_listed['corp_name'] == name].iloc[0,0]
  data = dart_fss.api.finance.fnltt_singl_acnt(corp_code, '2021', '11011')
  df = pd.DataFrame(data['list'])

  # fs_div가 CFS 이며 account_nm이 이익잉여금이라는 조건을 가진 것을 df 안에서 보기
  cond = (df['fs_div'] == 'CFS') & (df['account_nm']=='이익잉여금')
  df = df[cond]

  # 맨 뒤에 삼성전자 라는 column 추가하기
  df['name'] = name

  # 필요한 두개의 column만 골라내기
  df = df[['name','thstrm_amount','frmtrm_amount']]

  # column 이름 바꿔주기
  df.columns = ['기업명','당기','전기']

  # 당기와 전기의 문자를 숫자로 바꿔주자.
  df['당기'] = pd.to_numeric(df['당기'].str.replace(',',''))
  df['전기'] = pd.to_numeric(df['전기'].str.replace(',',''))

  # 증감 추가
  df['증감'] = df['당기'] - df['전기']
  # 전기에 비해서 증감이 얼마나 되었는가..? # 전기값은 음수도 가질 수 있기 때문에,, 율은 절대값이기 때문에 전기에 절대값의 명령어인 abs를 붙혀준다.
  df['증감율'] = abs(df['증감'])/abs(df['전기'])
  return df

get_profit('현대자동차')
```

```python
# 위의 함수를 받아 자동화

names = list(df_listed.sample(10)['corp_name'])

dfs = []

for name in names:
  try:
    df = get_profit(name)
    dfs.append(df)
  except:
    print(f'error-{name}')


df_result = pd.concat(dfs)

df_result

df_result.sort_values(by = '증감율', ascending = False)
```

```python
# 숙제

def get_salary(name):
  # 상장사중에 남여 차이가 적은 순서대로 정렬
  # corp_name, sexdstn, jan_salary_am 이 3가지의 정보가 필요함
  # 1) 상장사에서 기업이름을 가져와서 기업 코드를 변수로 두기
  corp_code = df_listed[df_listed['corp_name'] == name].iloc[0,0]
  # 2) 기업 코드를 통해 남여 연봉을 알기
  data = dart_fss.api.info.emp_sttus(corp_code, '2021', '11011')
  # 3) 남여 연봉을 가지고 dataframe으로 만들기
  df = pd.DataFrame(data['list'])
  df
   # 4) 필요한 columm 만 뽑아서 dataframe으로 만들기
  df = df[['corp_name', 'sexdstn', 'jan_salary_am']]

  df

  # 5) 여자의 평균 연봉을 남자의 평균 연봉 옆으로 가져오기
  # 다시 하나 만들기
  # document로 기업명은 카카오, 성별이 남자인 셀의 마지막번째 정보를 가져와라
  df_result = pd.DataFrame()
  doc = {
      '기업명' : name,
      '연봉(남)' : df[df['sexdstn'] == '남'].iloc[-1,-1],
      '연봉(여)' : df[df['sexdstn'] == '여'].iloc[-1,-1]
  }
  # 6) 솔직히 잘 모르겠음
  df_result = df_result.append(doc, ignore_index = True)


  # 7)각 연봉 정보를 숫자로 만들기
  df_result['연봉(남)'] = pd.to_numeric(df_result['연봉(남)'].str.replace(',',''))
  df_result['연봉(여)'] = pd.to_numeric(df_result['연봉(여)'].str.replace(',',''))
  df_result['남녀연봉차이'] = abs((df_result['연봉(남)']) - (df_result['연봉(여)']))

  return df_result
```

```python
# 숙제 자동화
# 기업 이름 list해 놓기
names = list(df_listed['corp_name'])


dfs = [] # 쌓을 list 만들기
for name in names:
  try:
    df = get_salary(name) # 함수 넣기
    dfs.append(df) # 함수 진행 시키며 계속 쌓기
  except:
    print(f'error-{name}') # 오류 나는 애들은 띄우기

df_result = pd.concat(dfs) # dfs 모두 합쳐서 보여주기

df_result.sort_values(by = '남녀연봉차이', ascending=True) # 제일 위에 있는애 보여주기
```
